## Dummy and standardise

>condition1 and condition2 describe the presence of a particular feature, here we have two categorical variables to account for the presence of multiple features. Normal is the absence of a feature. I will dummy these variables and add the corresponding features together, and drop the Normal columns (as they can be inferred by the absence of other features).

```python
#dummy and combine condition 1 and 2, which are the same variables but account for the presence of multiple
#categorical variables
df1=pd.get_dummies(df1, columns=['condition1','condition2'])
```

```python
#add instances in conition2 to corresponding column in condition1
df1.condition1_Artery=df1.condition1_Artery+df1.condition2_Artery
df1.condition1_Feedr=df1.condition1_Feedr+df1.condition2_Feedr
df1.condition1_PosA=df1.condition1_PosA+df1.condition2_PosA
df1.condition1_PosN=df1.condition1_PosN+df1.condition2_PosN
df1.condition1_RRAe=df1.condition1_RRAe+df1.condition2_RRAe
df1.condition1_RRAn=df1.condition1_RRAn+df1.condition2_RRAn
#df1.condition1_RRNe=df1.condition1_RRNe+df1.condition2_RRNe no instances of RRne in condition2 column
df1.condition1_RRNn=df1.condition1_RRNn+df1.condition2_RRNn
```

```python
#drop condition2 columns and norm category column from condition_1 
df1.drop(columns=['condition2_Artery','condition2_Feedr','condition2_Norm','condition1_Norm',
                 'condition2_PosA','condition2_PosN','condition2_RRAe','condition2_RRAn',
                 'condition2_RRNn'],axis=1, inplace=True)
```

## Create predictor and target dataframes

    Produce target variable (residuals), engineer new features, dummy categorical variables, train/test split, standardise

```python
#find residuals of all model 1 predictions and add to df2 as target column
all_predictions=model.predict(scaler.fit_transform(df1.drop(columns=['saleprice'])))
df2['target']=round(df1['saleprice']-all_predictions,2)
```

>bsmtfintype1 is a type of finish, bsmntfinsf1 is the square feet associated with that type of finish, the same is true of bsmtfintype2 and bsmtfinsf2. bsmtunfsf is the volume of unfinished basement, unfinished is a category in bsmtfintype1 and 2.<br><br>
    We saw in our first model that basement type was an important predictor, so our model is likely to improve if we can accurately note the relationship between these variables via feature engineering.<br><br>
    I will do this by dummying the variables, then multiplying the dummied variable by the corresponding sqft associated with that variable (so 1 would convert to number of square feet of that feature). We will then combine (with addition) columns that describe the same feature

```python
df2=pd.get_dummies(df2, columns=['bsmtfintype1','bsmtfintype2'])
```

```python
df2.columns
```

```python
#combine variables (don't run more than once!)

df2['bsmtfintype1_ALQ']=df2['bsmtfintype1_ALQ']*df2['bsmtfinsf1']
df2['bsmtfintype1_BLQ']=df2['bsmtfintype1_BLQ']*df2['bsmtfinsf1']
df2['bsmtfintype1_GLQ']=df2['bsmtfintype1_GLQ']*df2['bsmtfinsf1']
df2['bsmtfintype1_LwQ']=df2['bsmtfintype1_LwQ']*df2['bsmtfinsf1']
df2['bsmtfintype1_NA']=df2['bsmtfintype1_NA']*df2['bsmtfinsf1']
df2['bsmtfintype1_Rec']=df2['bsmtfintype1_Rec']*df2['bsmtfinsf1']
df2['bsmtfintype1_Unf']=df2['bsmtfintype1_Unf']*df2['bsmtfinsf1']

df2['bsmtfintype2_ALQ']=(df2['bsmtfintype2_ALQ']*df2['bsmtfinsf2'])+df2['bsmtfintype1_ALQ']
df2['bsmtfintype2_BLQ']=(df2['bsmtfintype2_BLQ']*df2['bsmtfinsf2'])+df2['bsmtfintype1_BLQ']
df2['bsmtfintype2_GLQ']=(df2['bsmtfintype2_GLQ']*df2['bsmtfinsf2'])+df2['bsmtfintype1_GLQ']
df2['bsmtfintype2_LwQ']=(df2['bsmtfintype2_LwQ']*df2['bsmtfinsf2'])+df2['bsmtfintype1_LwQ']
df2['bsmtfintype2_NA']=(df2['bsmtfintype2_NA']*df2['bsmtfinsf2'])+df2['bsmtfintype1_NA']
df2['bsmtfintype2_Rec']=(df2['bsmtfintype2_Rec']*df2['bsmtfinsf2'])+df2['bsmtfintype1_Rec']
df2['bsmtfintype2_Unf']=(df2['bsmtfintype2_Unf']*df2['bsmtfinsf2'])+df2['bsmtfintype1_Unf']+df2['bsmtunfsf']

#drop columns who'se features have been combined
df2.drop(columns=['bsmtfintype1_ALQ','bsmtfintype1_BLQ','bsmtfintype1_GLQ','bsmtfintype1_LwQ','bsmtfintype1_NA',
                 'bsmtfintype1_Rec','bsmtfintype1_Unf','bsmtunfsf','bsmtfinsf1','bsmtfinsf2'], inplace=True)
```

## Overall strategy based on models

The first model gave reliable predictions of sale price based on fixed charcteristics, with an R^2 value of .88 . on test data. There was a slight skew in the test residuals, we slightly undervalued properties on average (could be our model, might even be inflation!).

The second model did not give very reliable predictions on the effect of changeable characteristics, with a R^2 value of 0.17. It did however perform better than the baseline model, so would be a welcome addition to our first model when evaluating the price of a house for purchase based on all it's chacteristics. 

To implement these models in a business context, we would predict a price for a property for sale based on it's fixed and renovatable characteristics. We can compare the price of the property for sale with out predicted price to evaluate whether the property is over or under valued. If the property is under valued we can purchase it, and realise the difference in value as equity.

When deciding what aspects to renovate, we should examine the correlation coefficients in their standardised and dollar format to develop strategies on what changes to make. We can then query patricular proposals by evaluating the difference in predicted price of the property in its current state and with the proposed renovations, and compare that with the associated cost of those renovations to estimate the profit yielded with that strategy. We may want to, because of the models low R^2 value,  set a threshold of profitability to reduce our exposure to unprofitable outcomes.

If we had a comprehensive data on the cost of each type of renovation we could automate the process of querying strategies.

If we were particularly agressive we could even purchase overvalued properties if we could make up the difference and more in renovations, (taking a house out of the severley damaged category improves predicted value by 25k, if it only cost 5k to repair a 5k over valuation would be ok).

Some of the changeable characteristics are month sold and sale type, these are in theory low cost adjustments that we can make to our marketing strategy (for example our model tells us to selling in May is more profitable).


## Can we we use the model to inform investment decisions?

It all comes down to a question of variance. 

If the models residuals on test data are normally distrbuted (roughly are but not exactly!) and the model is built on data that is an accurate reflection of the market we are investing in, then we can use our model to inform our renovation decisions and know that in the long run we will come out ahead. The true dollar variance of this strategy could be high, so a large bankroll may be needed to reduce the risk of ruin. We can get an indication of this variance in the histogram of the models residuals, that show how far out in dollars we can expect to be from our predictions. 

The sum of all residuals of test data is positive, implying the test residuals are not normally distributed and that we have on average undervalued properties, this could be a costly mistake. It could be that variables outside our model (macro economic factors) are causing an increase/decrease on the global mean price of properties and introducing these into our model may reduce bias in (and improve) the test set results.

We also need to consider the 'real world' and constraints; to what extent does this snapshot of the market represent the wider market (the sample is quite small)? will this change over time? how much money does the business have to spend? how many opportunities for purchase are available? what are the predicted costs associated with specific renovations? All these will inform the extent to which the business can withstand the consequences of the model being inaccurate.

## Nearest neighbour model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
 X_train, X_test, y_train, y_test= train_test_split(df_m, target, test_size=.25, stratify=target)
```

```python
#standardise data 
x_col=X_train.columns
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train=pd.DataFrame(X_train, columns=x_col)
X_test=pd.DataFrame(X_test, columns=x_col)
```

```python
knn_params = {
    'n_neighbors': [1, 3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']}
knn = KNeighborsClassifier()
knn_gridsearch = GridSearchCV(knn,
                              knn_params,
                              n_jobs=2, 
                              cv=5, 
                              verbose=0, 
                              return_train_score=True)
```

```python
knn_gridsearch.fit(X_train, y_train)
```

```python
knn_gridsearch.best_params_
```

```python
best_knn = knn_gridsearch.best_estimator_
print(best_knn.score(X_train, y_train))
print(best_knn.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

predictions = best_knn.predict(X_test)
confusion = confusion_matrix(y_test, predictions, labels=[0, 1])
pd.DataFrame(confusion, columns=['predict_norm','predict_abnorm'],
            index=['is_norm','is_abnorm'])
```

```python
print(classification_report(target, predictions))
```

>The 'best model' in knn is equivelant to the baseline model, this is because the ranking is based on accuracy of predictions. This is a consequence of severe class imbalance.

>Given identifying true positives (abnormal sale real and predicted) is our main goal, we should conssider altering the probability threshold at which we classify our predictions. Even if our model becomes more inaccurate overall, we would benefit by having, for example, all abnormal sale outcomes correctly predicted (and a number of normal sales predicted as abnormal). This way we could say 'we can predict abnormal sales with X% accuracy', which is more helpful to us than a more accurate model that considers no sale abnormal.

```python
predict_p=best_knn.predict_proba(X_test)
```

```python
prediction_list=[]
for i in np.linspace(0,1,11):
    prediction_list.append([1 if p[1]>i else 0 for p in predict_p])
    
```

```python
w=0
for pred in prediction_list:
    confusion = confusion_matrix(y_test, pred, labels=[0, 1])
    if confusion[0][1]+confusion[1][1]==0:
        print('Probability threshold:', round(w,1),'\n no abnormal instances predicted\n' )
        w=w+.1
        
    else:
        print('Probability threshold:', round(w,1), '\n',confusion[0][1]+confusion[1][1],'abnormal instances predicted',
              round(confusion[1][1]/(confusion[0][1]+confusion[1][1]),3), 'correct prediciton rate')
        w=w+.1
        print(confusion,'\n')
              
```

>Our best accuracy, by changing the probability threshold was a 12.5% correct prediction rate on abnormal sales.

## Engineering lower class imbalance

The severe class imbalance has made it very difficult to correctly identify instances of the chosen class. To address this we are going to duplicate the set of houses that sold under abnormal circumstances. This works to overcome the effect of class imbalance when abnormal sales are clustered in a particular vector space but otherwise drowned out.

```python
#split train test
X_train, X_test, y_train, y_test= train_test_split(df_m, target, test_size=.3, stratify=target)
```

```python
#add target variable to train data
X_train=pd.concat([X_train, pd.DataFrame(y_train, columns=['target'])], axis=1)
```

```python
#double the instances of target variable in training data by duplicating them all
X_train=pd.concat([X_train,X_train[X_train.target==1]], axis=0)
```

```python
#remove target from x train and assign to y_train
y_train=X_train.pop('target')
```

```python
y_train.sum() #we have doubled the number of instances in class in training data
```

```python
#scale data
x_col=X_train.columns
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train=pd.DataFrame(X_train, columns=x_col)
X_test=pd.DataFrame(X_test, columns=x_col)
```

```python
knn_params = {
    'n_neighbors': [1, 3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']}
knn = KNeighborsClassifier()
knn_gridsearch = GridSearchCV(knn,
                              knn_params,
                              n_jobs=2, 
                              cv=5, 
                              verbose=0, 
                              return_train_score=True)
```

## Conclusion

Correctly predicting class when there are few instances of that class in a test set is a difficult task.

The first models did not perform well, with logistic regression being redundant when attempted with too many variables. 

By engineering lower class imbalance we saw an slight improvement in our ability to identify abnormal properties. To improve our model further we could explore this idea further. In our knn model we duplicated instances where the target class was 1, we could also randomly take away a set of instances where the target class was 0.

We could also investigate models further by removing some of the predictor variables.

Although the model is inaccurate it could still be useful in a business context; if we were faced with hundreds of thousands of properties where 6% of them were predicted an abnormal sale, reducing this to a smaller set where up to 12% are predicted an abnormal sale could be a low cost way of narrowing down on our target (though likely not useful given the particular problem posed in this question).
